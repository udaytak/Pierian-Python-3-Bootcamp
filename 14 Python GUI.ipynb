01 - Interact and Interactive

```python
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func,x=10)
```

```python
interact(func,x=True)
```

```python
interact(func,x="Hello")
```

```python
@interact(x=True,y=1.0)
def g(x,y):
    return  (x,y)
```

```python
@interact(x=True,y=fixed(1.0))
def g(x,y):
    return  (x,y)
```

```python
interact(func, x=1)
```

```python
interact(func, x=10)
# min = x, max = x*3
```

```python
interact(func, x=widgets.IntSlider(min=-100,max=100,step=1,value=0))
```

```python
interact(func, x=(-10,10,1))
```

```python
interact(func, x=(-10.0,10.0,0.1))
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(func, x='hello')
```

```python
interact(func, x=['hello','Opt2','Opt3']) #list
```

```python
interact(func, x={'one':10,'two':20}) #dictinary
```

```python
interact(func, x={'one':10,'two':20}); #adding  semicolon in the end
```

```python
#link 2 or more widgets together
a = widgets.FloatText()
b = widgets.FloatSlider()

display(a,b)

mylink = widgets.jslink((a,'value'),(b,'value'))
```

```python
#link 2 or more widgets together
a = widgets.FloatText()
b = widgets.FloatSlider()

display(a,b)

mylink = widgets.jslink((a,'value'),(b,'max')) #you can link to max
```

```python
mylink.unlink() #unlinking
```
